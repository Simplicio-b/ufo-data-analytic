```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import plotly.express as px
```

```python
sns.color_palette("RdPu")
```

```python
df3= pd.read_csv("./df3.csv")
```

```python
df3.info()
```

Podemos trabalhar com o histograma ou com o barplox, vamos interpretar os dados e saber qual se adéqua a cada coluna:

```python
plt.figure(figsize=(10,5))
sns.barplot(x= df3.value_counts(df3["mday"]).index, y = df3.value_counts(df3["mday"]).values, palette = sns.color_palette("RdPu"))
plt.xlabel('Dias')
plt.ylabel('Numeros de Registros')
plt.title('Registros de UAP por dia(1997-2017)');
```

```python
sns.barplot(x= df3.value_counts(df3["month"]).index, y = df3.value_counts(df3["month"]).values, palette = sns.color_palette("RdPu"))
plt.xlabel('Meses')
plt.ylabel('Numeros de Registros')
plt.title('Registos de UAP por Meses(1997-2017)');
```

```python
sns.barplot(x= df3.value_counts(df3["year"]).index, y = df3.value_counts(df3["year"]).values, palette = sns.color_palette("RdPu"))
plt.xlabel('Anos')
plt.xticks(rotation= 90)
plt.ylabel('Numero de Registros')
plt.title('Registros de UAP por Ano(1997-2017)');
```

Agora temos uma visualização melhor dos nossos dados referente aos dias, meses e anos.

```python
sns.histplot(df3['hour'], kde = 'True', color="crimson")
plt.title('Horario dos avistamentos de UAP (1997-2017)')
plt.xlabel('Hora')
plt.ylabel('Registros de UAP')
```

Para as horas eu usei o histograma, histograma é basicamente um gráfico de barras para variaveis quatitativas, que são dividias em intervalos, e nos mostra a frequencia de dados que se tem em cada intervalo, o Kde é uma estimativa da densidade desses dados. Achamos interessante mostrar com as horas pois é possivel observar uma alta densidade nos horarios da noite.

```python
plt.figure(figsize=(10,5))
sns.barplot(x= df3.value_counts(df3["wdire"]).index, y = df3.value_counts(df3["wdire"]).values, palette = sns.color_palette("RdPu"))
plt.xlabel('Orientação')
plt.xticks(rotation=90)
plt.ylabel('Numeros de Registros')
plt.title('Registros de UAP por localização Geografica(1997-2017)');
```

```python
plt.figure(figsize=(7,3))
plt.xticks(fontsize=8)
plt.xticks(rotation=80)
sns.countplot(x='state',order=df3['state'].value_counts().index,data=df3)
plt.xlim(-0.5,9.5)
plt.title('Registros de UAP por Estados(1997-2017)')
plt.xlabel('Estados')
plt.ylabel('Registros de UAP')
```

como o dataframe apresenta muitos valores unicos, como nas colunas cidade, estado e condições de tempo na hora das aparições, optei pelo grafico countplot, pois ele apresenta a propiedade de podermos escolher os n primeiros dados com maior valor.

```python
plt.figure(figsize=(8,4))
plt.xticks(fontsize=8)
plt.xticks(rotation=80)
plt.title('10 Cidades com mais registros de UAP(1997-2017)')
plt.xlabel('Cidades')
plt.ylabel('Registros de UAP')
sns.countplot(x='city',order=df3['city'].value_counts().index,data=df3)
plt.xlim(-0.5,9.5)
```

```python
plt.figure(figsize=(7,3))
plt.xticks(fontsize=6)
plt.xticks(rotation=80)
sns.countplot(x='conds',order=df3['conds'].value_counts().index,data=df3)
plt.title('Condições do tempo(1997-2017)')
plt.xlabel('Condições')
plt.ylabel('Registros de UAP')
plt.xlim(-0.5,9.5)
```

Os 4 graficos acima apresentam dados geograficos dos locais de aparições que vão ser utilizados em uma mapa no próximo notebook para uma observação mais detalhada. O ultimo grafico apresenta informações das condições climaticas na hora das observações, seram muito relevante para nossas proximas etapas.

```python
plt.figure(figsize=(7,5))
sns.barplot(x= df3.value_counts(df3["shape"]).index, y = df3.value_counts(df3["shape"]).values, palette = sns.color_palette("RdPu"))
plt.xlabel('Morfologia')
plt.xticks(fontsize=8)
plt.xticks(rotation=70)
plt.ylabel('Numeros de Registros')
plt.title('Caracteristicas de UAP-Forma(1997-2017)');
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x= df3.value_counts(df3["colour"]).index, y = df3.value_counts(df3["colour"]).values, palette = sns.color_palette("RdPu"))
plt.xlabel('cor')
plt.xticks(fontsize=7)
plt.xticks(rotation=85)
plt.ylabel('Numeros de Registros')
plt.title('Caracteristicas dos UAP-Cor(1997-2017)');
```

Os dois ultimos graficos apresentam caracteristicas morfologicas dos UAP, tambem seram muito relevantes nas nossas interpretações para responder algumas questões. 

O grafico de dispersão ou Scatter plot são representações gráficas do relacionamento entre duas variáveis numéricas. O Scatter plot utiliza pontos para representar essa relação, cada ponto representa o valor de uma variável no eixo horizontal e o valor de outra variável no eixo vertical. Vamos usar esse grafico para vê as relaçoes das colunas climaticas do nosso dataframe.

Logo abaixo fizemos um grafico de dispersão dos dias, meses e anos em relaçaõ a temperatura. Observamos dados normais, com alguns outlier, mas sem muita relavancia já que estamos falando de temperatura em relação ao tempo. Por exemplo no gráfico dos meses é possivel observar a variação da temperatura com os meses, explicando as estações do ano.

```python
grafico = sns.scatterplot(data=df3, x="mday", y="temp")
grafico.set_title('Dias X Temperatura')
grafico.set_xlabel('Dia', fontsize=12)
grafico.set_ylabel('Temperatura', fontsize=12)
```

```python
grafico2=sns.scatterplot(data=df3, x="month", y="temp")
grafico2.set_title('Meses X Temperatura')
grafico2.set_xlabel('Meses', fontsize=12)
grafico2.set_ylabel('Temperatura', fontsize=12)
```

```python
grafico3 = sns.scatterplot(data=df3, x="year", y="temp")
grafico3.set_title('Ano X Temperatura')
grafico3.set_xlabel('Ano', fontsize=12)
grafico3.set_ylabel('Temperatura', fontsize=12)
```

O grafico de dispersão é muito importante para algumas visualizaçãoes, mas para nós não fez muito sentido pelo nossos tipos de dados
