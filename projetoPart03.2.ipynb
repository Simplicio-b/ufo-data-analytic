Nós resolvemos escolher um dataset especifico para trabalharmos, escolhemos o que tem mais dados do nosso interesse. Tambem realizamos, previamente, a limpeza do dataframe para remover colunas sem valores e linhas duplicadas.

A função .describe() ela, por padrão, escolhe as colunas com dados quantitativos e tem propiedades parecidas com o boxplot, mas com uma visualização diferente de um grafico. Usamos essa função para termos detalhes das nossas colunas, fazer a melhor analise e escolher o melhor grafico para ter uma visualização adequada.

```python
df.describe() 
```

A linha count tem a contagem de valores de cada coluna, com isso conseguimos entender o tamanho da amostra. Inclusive conseguimos detectar linhas com valores vazios comparando o valor do count entre duas colunas.
Mean é a média e std é o desvio padrão.
Min e max são os valores que nos auxiliam para identificar a amplitude da amostra. 
Os valores de 25%, 50% e 75% são os quartis, são os valores que nos mostram de que forma os dados foram distribuídos.


Podemos usar dados categoricos com o .describe(), desde que os especificamos as colunas na entrada. Como os dados são categóricos, a função describe retorna cálculos mais adequados para estes tipos de dados, como a contagem de valores distintos (unique), a moda (top) e a frequência da mesma (freq), assim como tambem é possivel saber quanto de valores vazios cada coluna tem.
Com essa visualizão nós conseguimos pensar em informações que queremos extrair desse dataframe e, o que nos interessa nessa parte do projeto, atraves de qual grafico fica melhor a visualização das informações.

```python
df[["state", "city", "colour", "shape", "wdire", "conds"]].describe()
```

Fiz uso de duas formas para criar os graficos, um deles foi o histograma, de inicio ia usar apenas ele, mas como o dataframe apresenta muitos valores unicos, como nas colunas cidade, estado e condições de tempo na hora das aparições, optei pelo grafico countplot, pois ele apresenta a propiedade de podermos escolher os n primeiros dados com maior valor, sendo assim por exemplo posso apresentar as 10 cidades com mais aparições ja que seria impossivel colocar as 8.845 cidades unicas que o dataframe apresenta.

```python
sns.histplot(df['mday'], kde = 'True').set_title ('Aparições de OVINs por dia')
```

```python
sns.histplot(df['month'], color = 'black', kde = 'True').set_title ('Aparições de OVINs por Mês')
```

```python
plt.figure(figsize=(7,3))
plt.xticks(rotation=80)
sns.countplot(x='year', data=df).set_title ('Aparições de OVINs por ano')
```

```python
sns.histplot(df['hour'], color = 'blue', kde = 'True').set_title ('Aparições de OVINs por Hora')
```

O grafico destaca que as aparições são recorrentemente vistas a noite.

Logo abaixo os graficos das colunas com dados geograficos estado e região:

```python
plt.figure(figsize=(7,3))
plt.xticks(fontsize=8)
plt.xticks(rotation=80)
sns.countplot(x='state',order=df['state'].value_counts().index,data=df).set_title ('Dez Estado com mais aparições de OVINs')
plt.xlim(-0.5,9.5)
```

```python
sns.histplot(df['wdire'], color = 'pink').set_title ('Aparições de OVINs por Região')
plt.xticks(rotation='vertical')
plt.xticks(fontsize=7)
plt.figure(figsize=(5,5))
```

```python
sns.histplot(df['shape'], color = 'orange').set_title ('Formato dos OVINs avistados')
plt.xticks(rotation=75)
plt.xticks(fontsize=8)
plt.figure(figsize=(2,2))
```

```python
sns.histplot(df['colour'], color = 'violet').set_title ('Cor dos OVINs avistados')
plt.xticks(rotation=80)
plt.xticks(fontsize=6)
plt.figure(figsize=(7,3))
```

```python
plt.figure(figsize=(7,3))
plt.xticks(fontsize=6)
plt.xticks(rotation=80)
sns.countplot(x='conds',order=df['conds'].value_counts().index,data=df).set_title ('Condições de tempo na hora das aparições')
plt.xlim(-0.5,9.5)
```

```python
plt.figure(figsize=(8,4))
plt.xticks(fontsize=8)
plt.xticks(rotation=80)
sns.countplot(x='city',order=df['city'].value_counts().index,data=df).set_title ('As dez cidades com mais aparições de OVINs')
plt.xlim(-0.5,9.5)
```

Uma observação feita é que foram colocadas colunas com dados das condições climaticas das regiões na hora das aparições como as colunas snow(neve), tornado, fog(fogo), thunder(torvão) e rain(raios) que tem valores booleanos, mas esses mesmo dados foram especificados na coluna conds, por isso ela ficou muitas com muitas variaveis unicas, sendo assim para não ficar redundante não apresentei o grafico dessas colunas, tambem vamos analisar se removemos elas do dataframe ou removemos elas da coluna conds.
