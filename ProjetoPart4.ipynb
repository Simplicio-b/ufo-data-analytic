O objetivo dessa etapa é aplicar o K-means nos conjuntos de dados usados no projeto para descobrir possíveis grupos.

Vamos utilizar do grafico de dispersão para aplicar o kmeans

```python
import numpy as np
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import plotly.graph_objects as go
```

```python
from sklearn.cluster import KMeans
```

Analisamos nossos dataframes e observamos que o dataframe indicado para essa etapa é o primeiro, com os dados da latitude e longitude. Podiamos fazer com o segundo, mas ele tem muito mais dados e assim os pontos ficariam muito distante impossibilitado as analises. 

```python
df1= pd.read_csv("./df1.csv")
```

```python
n = np.array(df1[["lat", "lng",]])
```

```python
kmeans = KMeans(n_clusters = 4, init = 'random',n_init = 10, max_iter = 100)
pred_y = kmeans.fit_predict(n)
plt.scatter(n[:,1], n[:,0], c = pred_y)
plt.xlim(-165, -60)
plt.ylim(0, 70)
plt.grid()
plt.scatter(kmeans.cluster_centers_[:,1], kmeans.cluster_centers_[:,0], s =70, c = "blue")
```

Nós optamos por 4 centroides pela observação das disposições dos pontos. Aplicamos o random para gerar aleatoriamente a localização dos centroides. Observamos que apesar da aletoriedade existe uma padrão. A ideia que tivemos é que cada centroide podeia ser visto como uma possivel posto de observação, deixaremos essa ideia mais madura no relatorio.
