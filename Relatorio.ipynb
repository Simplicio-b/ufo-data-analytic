E coincidentemente o Escritório de Resolução de Anomalias (AARO) da NASA, criado em 2022, juntamente com a sede do Departamento de Defesa dos Estados Unidos,Pentágono, tornou públicos relatórios sobre UAP e uma [Site](https://www.aaro.mil/) que apresenta informações sobre casos resolvidos, não resolvidos e alguns vídeos mostrando os OVNIs.
